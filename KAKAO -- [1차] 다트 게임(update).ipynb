```python
def star(score) :    
    score[-1] *= 2
    score[-2] *= 2
    
    return score

def sharp(score) :
    score[-1] *= -1
    
    return score

def double(score) :
    score[-1] = score[-1]**2
    
    return score

def triple(score) :
    score[-1] = score[-1]**3
    
    return score

def solution(dartResult) :
    score = [0]
    number = {}
    
    for i in range(11) :
        number[str(i)] = i
        
    for idx, char in enumerate(dartResult) :
        if char in number :
            if char == '0' and dartResult[idx-1] in number :
                score[-1] = number[dartResult[idx-1]+'0']
            else :
                score.append(number[char])
        elif char == '*' :
            star(score)
        elif char == '#' :
            sharp(score)
        elif char == 'D' :
            double(score)
        elif char == 'T' :
            triple(score)
            
    return sum(score)
```
